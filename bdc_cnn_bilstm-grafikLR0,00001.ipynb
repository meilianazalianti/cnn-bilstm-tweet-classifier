## Menghapus Teks di dalam Tanda Kurung Siku

```python
import re
# Menghapus teks di dalam tanda kurung siku []
def remove_text_inside_brackets(text):
    # Menghapus teks di dalam tanda kurung siku
    text = re.sub(r'\[.*?\]', '', text)
    return text

# Contoh penggunaan dengan kolom 'text' pada DataFrame
df['text'] = df['text'].apply(lambda x: remove_text_inside_brackets(x))
df
```

### Remove karakter khusus & url

```python
# Fungsi pra-pemrosesan teks: Menghapus angka, karakter khusus, dan URL
def remove_numbers_and_urls(text):
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Menghapus URL
    text = re.sub('[^A-Za-z ]+', '', text)
    text = re.sub(r'²', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r"\W", " ", text)
    return text

# Menerapkan fungsi ke kolom 'text' pada dataframe
df['text'] = df['text'].apply(lambda x : remove_numbers_and_urls(x))
df
```

## Menghapus Huruf Acak yang Panjangnya Lebih dari 15 Karakter

```python
# Fungsi untuk menghapus huruf acak yang panjangnya lebih dari 15 karakter
def remove_random_letters(text):
    # Menghapus huruf acak yang panjangnya lebih dari 15 karakter
    text = ' '.join([word for word in text.split() if len(word) <= 15])
    return text

# Menerapkan fungsi ke kolom 'text' pada dataframe
df['text'] = df['text'].apply(lambda x: remove_random_letters(x))
df
```

## Pembagian Data

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_pad, y, test_size=0.1, random_state=42)
```

```python
from keras.utils import to_categorical

# Mengubah label menjadi kategori one-hot
y_train = to_categorical(y_train, num_classes=8)
y_val = to_categorical(y_val, num_classes=8)
```

```python
X_train = np.array(X_train)
X_val = np.array(X_val)
y_train = np.array(y_train)
y_val = np.array(y_val)
```

## cnn+bilstm

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Conv1D, MaxPooling1D, LSTM, Bidirectional, Dense, Dropout, GlobalMaxPooling1D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score, classification_report
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
from tensorflow.keras.models import Sequential

# Membuat model Sequential
model_cnn_bilstm = Sequential()

# Menambahkan layer Embedding
model_cnn_bilstm.add(Embedding(len(word_vectors.key_to_index), 
                               100, 
                               weights=[word_vectors.vectors], 
                               input_length=100, 
                               trainable=False))

# Menambahkan layer Conv1D
model_cnn_bilstm.add(Conv1D(filters=128, kernel_size=5, activation='relu'))

# Menambahkan layer MaxPooling1D
model_cnn_bilstm.add(MaxPooling1D(pool_size=3))

# Menambahkan layer BiLSTM
model_cnn_bilstm.add(Bidirectional(LSTM(64, return_sequences=False)))

# Menambahkan layer Dense
model_cnn_bilstm.add(Dense(8, activation='softmax'))

# Kompilasi model
model_cnn_bilstm.compile(optimizer=Adam(learning_rate=0.00001), loss='categorical_crossentropy', metrics=['accuracy'])

# Menampilkan ringkasan model
model_cnn_bilstm.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss', mode = 'min', patience=10, restore_best_weights=True)
    
history = model_cnn_bilstm.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_val, y_val), callbacks=[early_stopping])
```

```python
# Evaluasi model CNN+BiLSTM
loss, accuracy = model_cnn_bilstm.evaluate(X_val, y_val)
print('Accuracy: {:.2f}%'.format(accuracy * 100))
print('Loss: {:.2f}'.format(loss))
```

```python
train_loss = history.history['accuracy']
val_loss = history.history['val_accuracy']

epochs = range(1, len(train_loss) + 1)
plt.plot(epochs, train_loss, 'green', label='Training accuracy')
plt.plot(epochs, val_loss, 'orange', label='Validation accuracy')
plt.title('accuracy Train dan Validasi')
plt.xlabel('Epochs')
plt.ylabel('accuracy')
plt.legend()
plt.savefig("plot_eval_acc_setelah.jpg")
plt.show()
```

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(train_loss) + 1)
plt.plot(epochs, train_loss, 'green', label='Training Loss')
plt.plot(epochs, val_loss, 'orange', label='Validation Loss')
plt.title('Loss Train dan Validasi')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig("plot_eval_loss_setelah.jpg")
plt.show()
```
